2. Para los atributos categ칩ricos cuente el n칰mero y despliegue la lista de categor칤as 칰nicas poratributo e indique si el atributo corresponde a un dato nominal ordinal. 


Se va a aplicar codigo para ver las categorias unicas en las columnas de atributos categoricos y ver si son nominales u ordinales.


```python
# Columnas categ칩ricas de inter칠s
columnas_categoricas = [2, 3, 4, 6, 7, 8, 17]

# Diccionario para almacenar los resultados
resultados = {}

# Recorrer las columnas categ칩ricas
for columna in columnas_categoricas:
    # Obtener las categor칤as 칰nicas y contar su n칰mero
    categorias_unicas = datos[columna].unique()
    num_categorias = len(categorias_unicas)
    
    # Determinar si el atributo es nominal u ordinal
    tipo_atributo = "Nominal" if num_categorias > 10 else "Ordinal"
    
    # Almacenar los resultados en el diccionario
    resultados[columna] = {"Numero de categorias": num_categorias, "Categorias unicas": categorias_unicas, "Tipo de atributo": tipo_atributo}

# Imprimir los resultados
for columna, info in resultados.items():
    print(f"Columna {columna}:")
    print(f"N칰mero de categor칤as: {info['Numero de categorias']}")
    print("Categor칤as 칰nicas:")
    for categoria in info['Categorias unicas']:
        print(categoria)
    print(f"Tipo de atributo: {info['Tipo de atributo']}")
    print("--------------------")



```

3. Codifique todos los atributos categ칩ricos, seleccione el m칠todo de codificaci칩n y explique porqu칠
selecciona cada m칠todo

Para la columna 2, se va autulizar el metodo ordinal. porque conserva la relaci칩n de orden entre las categor칤as de manera m치s compacta que la codificaci칩n One-Hot, lo que puede ser 칰til cuando la jerarqu칤a de las categor칤as es relevante para el an치lisis o el modelo de aprendizaje autom치tico.

En la columna dos, debido que es binaria se va a aplicar el metodo de codificacion one-hot, ya que tiene la capacidad de mostrar las categorias de manera no sesgada.

En la columna 4, se aplicara igual one - hot ya que es una columna binaria.

Columna 6, se va autulizar el metodo ordinal. porque conserva la relaci칩n de orden entre las categor칤as de manera m치s compacta que la codificaci칩n One-Hot, lo que puede ser 칰til cuando la jerarqu칤a de las categor칤as es relevante para el an치lisis o el modelo de aprendizaje autom치tico.

Igual para la columna 7.


```python

# columna 2

#Metodo ordinal
#  codificaci칩n ordinal
codificacion_ordinal = {
    'yoyota': 1,
    'nissan': 2,
    'mazda': 3,
    'mitsubishi': 4,
    'honda': 5,
    'volkswagen' : 6,
    'subaru' : 7,
    'peaugot' : 8,
    'volvo' : 9,
    'dodge' : 10,
    'mercedes-benz' : 11,
    'bmw' : 12,
    'audi' : 13,
    'plymouth' : 14,
    'saab' : 15,
    'isuzu' : 16,
    'jaguar' : 17,
    'chevrolet' : 18, 
    'alfa-romero' : 19,
    "renault" : 20,
    'mercury' : 21
}

datos['Columna2_codificada'] = datos[2].map(codificacion_ordinal)

# Mostrar el resultado
print(datos[[2, 'Columna2_codificada']].head())

```

```python
#Columna 3, Diesel o GAS

import pandas as pd


# Crear columnas binarias utilizando One-Hot encoding
datos['gas'] = datos[3].apply(lambda x: 1 if x == 'gas' else 0)
datos['diesel'] = datos[3].apply(lambda x: 1 if x == 'diesel' else 0)

# Verificar los cambios
print(datos[[3, 'gas', 'diesel']].head())
```

```python
#Columna 4, turbo o standar


# Crear columnas binarias utilizando One-Hot encoding
datos['standard'] = datos[4].apply(lambda x: 1 if x == 'std' else 0)
datos['turbo'] = datos[4].apply(lambda x: 1 if x == 'turbo' else 0)

# Verificar los cambios
print(datos[[4, 'standard', 'turbo']].head())

```

```python
#Columna 6
#Metodo ordinal
#  codificaci칩n ordinal
codificacion_ordinal = {
    'sedan': 1,
    'hatchback': 2,
    'wagon': 3,
    'hardtop': 4,
    'convertible': 5
}

datos['Columna6_codificada'] = datos[6].map(codificacion_ordinal)

# Mostrar el resultado
print(datos[[6, 'Columna6_codificada']].head())


```

```python
#Columna 7
#Metodo ordinal

# Codificaci칩n ordinal
codificacion_ordinal = {'fwd': 1, 'rwd': 2, '4wd': 3}

datos['Columna7_codificado'] = datos[7].map(codificacion_ordinal)

print(datos[[7, 'Columna7_codificado']].head())
```

```python
# Codificaci칩n one-hot de la columna 7
# Crear columnas binarias utilizando One-Hot encoding
datos['front'] = datos[8].apply(lambda x: 1 if x == 'front' else 0)
datos['rear'] = datos[8].apply(lambda x: 1 if x == 'rear' else 0)

# Verificar los cambios
print(datos[[8, 'standard', 'turbo']].head())
```

```python
#Columna 17


mapeo = {categoria: indice for indice, categoria in enumerate(datos.iloc[:, 17].unique())}

datos['Columna17_codificada'] = datos.iloc[:, 17].map(mapeo)

print(datos[[17, 'Columna17_codificada']].head())

```
